```python
#Showing rules and guidance
print('Rules reminder: for two players, X and O, who take turns marking the spaces in a 3×3 grid. The player who succeeds in placing three of their marks in a horizontal, vertical, or diagonal row is the winner.')

print('You will see the Board in a second. To indicated the cell to which you want to move you need to type number from 1 to 9, corresponding to your choice.')

import numpy as np
example = np.matrix([[7, 8, 9], [4, 5, 6], [1, 2, 3],])

print('The cells numbers on the board are as follows:')
print(example)


#Defining the Board. Board contains empty spaces which will be replaced with players turn: 
Board = {'7': ' ' , '8': ' ' , '9': ' ' ,
         '4': ' ' , '5': ' ' , '6': ' ' ,
         '1': ' ' , '2': ' ' , '3': ' ' }


board_keys = []

for key in Board:
    board_keys.append(key)

#Print Board. This function will be called after each turn to display updated board 

def printBoard(board):
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-+-+-')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-+-+-')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])

# Game function
def game():

    turn = 'X'
    count = 0
    player = 'Player_1'


    for i in range(10):
        printBoard(Board)
        print("Now it's turn of:" + player + ' with ' + turn + " sign. To which position on the board do you want to move?") 
        
        move = input()
        

        if Board[move] == ' ':
            Board[move] = turn
            count += 1
        else:
            print("That place is already filled.\nMove to which place?")
            continue
        
        
        # Winning combination will be achieved after 5 turns. So checking if player X or O has won,for every move after 5 moves for every moving combination. 
        if count >= 5:
            
            # across the top
            if Board['7'] == Board['8'] == Board['9'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")                
                break
            # across the middle    
            elif Board['4'] == Board['5'] == Board['6'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break
            # across the bottom    
            elif Board['1'] == Board['2'] == Board['3'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break
            # down the left side    
            elif Board['1'] == Board['4'] == Board['7'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break
            # down the middle    
            elif Board['2'] == Board['5'] == Board['8'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break
            # down the right side    
            elif Board['3'] == Board['6'] == Board['9'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break 
            # diagonal    
            elif Board['7'] == Board['5'] == Board['3'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break
            # diagonal    
            elif Board['1'] == Board['5'] == Board['9'] != ' ': 
                printBoard(Board)
                print("\nGame Over.\n")                
                print(" **** " + player + ' with ' +turn + " won. ****")
                break 

        # If neither X nor O wins and the board is full, we'll declare the result as 'tie'.
        if count == 9:
            print("\nGame Over.\n")                
            print("It's a Tie!!")

        # Swap the players after every move:
        if turn =='X':
            turn = 'O'
            player = 'Player_1'
        else:
            turn = 'X'
            player = 'Player_2'
    
    # Now we will ask if player wants to restart the game or not.
    restart = input("Do want to play Again?(y/n)")
    if restart.lower() == "y":  
        for key in board_keys:
            Board[key] = " "

        game()

game()
```
